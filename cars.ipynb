```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "second_cars_info.csv"
data = pd.read_csv(path,encoding='gbk')
data
```

```python
#删除未上牌的数据
data.drop(index=(data[(data.Boarding_time == '未上牌')].index),inplace=True)
```

```python
data[(data.Boarding_time == '未上牌')].index
```

```python
data['year'] = data.Boarding_time.str[:4].apply(lambda x:int(x))
```

```python
data.year
```

```python
data['month']=data.Boarding_time.str[5:-1].apply(lambda x:int(x))
```

```python
data
```

```python
#17 10 
data['diff_month'] = (2017-data.year) * 12 + (10-data.month)
```

```python
data.diff_month
```

```python
data.head(5)
```

```python
data["Km"] = data.Km.str[:-3]
```

```python
#data.Km.apply(lambda x:float(x))
index = data[~data.Km.str.contains("百")].index
data = data.loc[index,:]
```

```python
data[data.Km.str.contains("百")].index
data["Km_new"] = data.Km.astype("float")
```

```python
#1.查看二手车价格分布区间
min = data.Sec_price.min()
max = data.Sec_price.max()
```

```python
bins = [min,3,5,10,15,20,30,40,50,60,max]
result1 = pd.cut(data.Sec_price,bins)
```

```python
data.Sec_price.head(5)
```

```python
result1_show = result1.value_counts()
result1_show = result1_show.sort_index()
```

```python
result1_show.index
```

```python
X = np.arange(len(result1_show))
Y = result1_show.values
plt.figure(figsize=(12,6))
plt.bar(X,Y)
plt.xticks(X,result1_show.index,rotation=45)
#计算每一个条柱数据的百分比
precents = [str(round(i*100,2))+"%" for i in result1_show/result1_show.sum()]
for x,y,z in zip(X,Y,precents):
    plt.text(x-0.3,y+50,z)
```

```python
[str(round(i*100,2))+"%" for i in result1_show/result1_show.sum()]
```

```python
#使用气泡图 完成 汽车二手车价格和 行驶公里数 行驶时长的 关系
#plt.scatter(X,Y,R,C)
#奥迪 Audi
#大众 Das
#X 公里数
#Y 时长
#R 价格
some_cars = data[data.Brand.isin(["奥迪",'迈巴赫','大众',"别克"])]
colors = {
    "奥迪":'r',
    '大众':'orange',
    '别克':'g',
    '迈巴赫':'#ff00ff'
}
#plt.scatter(some_cars['Km_new'],some_cars["diff_month"],some_cars.Sec_price*2)
for brand in ["奥迪"]:#["奥迪",'迈巴赫','大众',"别克"]:
    temp = some_cars[some_cars.Brand == brand]
    plt.scatter(temp['Km_new'],\
                temp["diff_month"],\
                temp.Sec_price*2,
               color=colors[brand])
for brand in ["奥迪"]:#["奥迪",'迈巴赫','大众',"别克"]:
    temp = some_cars[some_cars.Brand == brand]
    plt.scatter(temp['Km_new'],\
                temp["diff_month"],\
                temp.Sec_price*2,
               color=colors[brand])
```

二手车价格主要影响因素有：使用时长，公里数，新车价格。
总的来说，使用时长越短，公里数越小，新车价格越贵，二手车价格就越贵。

```python
data['New_price_new'] = data.New_price.str[:-1].apply(lambda x:float(x))
data[['Sec_price','diff_month','Km_new','New_price_new']]
```

```python
data[['Sec_price','diff_month','Km_new','New_price_new']].corr()
```
