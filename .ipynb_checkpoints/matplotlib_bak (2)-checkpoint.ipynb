```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def listplot():
    #根据数据绘制折线图
    plt.figure()#产生fig对象，axic对象，
    #X = np.arange(5) #[0,1,2,3,4]
    #Y = np.arange(5)+1 #[1,2,3,4,5]  =>(0,1),(1,2)....
    X = np.linspace(0,10,100)
    Y = np.sin(X)
    plt.plot(X,Y,'-')
```

```python
def listplot2():
    plt.figure(figsize=(6,4))
    X = np.arange(0,5,0.5)
    Y = X**2
    #x [...] y[...] -->[(x,y),(x,y)]
    #x [1,2,3],y[2,3,4]-zip->[(1,2),(2,3),(3,4)]
    plt.plot(X,Y,'-',color='red')   
    #plt.text(5,100,"important")
    #for i,j in zip(X,Y):
    #    plt.text(i+0.3,j,j,bbox=dict(facecolor='red', alpha=0.5))
    plt.title("这个是可视化")
    plt.xlabel("x轴标签")
    plt.ylabel("y轴标签")
    plt.xticks(np.arange(0,5,1),['a','b','c','d','e'])
listplot2()
```

```python
def listplot3():
    X = np.arange(50)
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'-',color='r',label='a')
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'--',color='y',label='b')
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'-*',color='c',label='c')
    plt.legend(loc='upper right')#图示
listplot3()
```

```python
def listplot4():
    X = np.linspace(-np.pi,np.pi,256,endpoint=False)
    Y1,Y2 = np.sin(X),np.cos(X)
    plt.plot(X,Y1,'-',label='sin')
    plt.plot(X,Y2,'-',label='cos')
    plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],\
               [r'$\pi$',r'$-\pi/2$',r'0',r'$\pi/2$',r'$\pi$'])
    plt.legend()
    plt.ylim(-1,1)
    plt.xlim(-4,4)
    
    ax = plt.gca()# get current axic
    ax.spines["top"].set_color(None)
    ax.spines["right"].set_color(None)
    ax.spines['left'].set_position(('data',0))
    ax.spines['bottom'].set_position(('data',0))
    
    t = 2*np.pi/3
    y = np.sin(t)
    
    plt.scatter(t,y,50,color='r')
    plt.plot([t,t],[y,0],'--',color='r')
    #xycoords 表示当前的箭头的数据轴是基于整体数据的坐标轴
    plt.annotate("y=sin(x)",xy=(t,y),\
                 xycoords='data',textcoords='offset points',\
                 xytext=(30,30),\
                 arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.3'))
listplot4()
```

----------
x : sequence of scalars
    the x coordinates of the bars.

    *align* controls if *x* is the bar center (default) or
    left edge.

height : scalar or sequence of scalars
    the height(s) of the bars

width : scalar or array-like, optional
    the width(s) of the bars
    default: 0.8

bottom : scalar or array-like, optional
    the y coordinate(s) of the bars
    default: None

```python
#柱状图
X = np.arange(4)
GDP = [2000,2345,2567,1890]
plt.bar(X,GDP)
plt.xticks(np.arange(4),['bj','sh','sz','dl'])
plt.ylim(1500,2900)
```

```python
Y = np.arange(4)
GDP = [2000,2345,2567,1890]
plt.barh(Y,GDP)
plt.yticks(np.arange(4),['bj','sh','sz','dl'])
plt.xlim(1500,2900)
```
