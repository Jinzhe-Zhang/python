```python
#2.分析每一种英雄每个属性平均值
data["英雄类型"].value_counts()
print (data.groupby('英雄类型').mean().\
apply(lambda x:round(x,2)).\
sort_values('物理攻击'))
#近战英雄和远程英雄的数量
print (data['攻击范围'].value_counts())
print (data.groupby('攻击范围').count())
print (data.groupby('攻击范围')['英雄名称'].count())
#每一种英雄的生命回复的排名
print (data.groupby('英雄类型')['生命回复'].mean()\
.apply(lambda x:round(x,2)).\
sort_values(ascending=False))
#孙悟空 初始攻击力
temp=data.sort_values('物理攻击',ascending=False)
temp.index = np.arange(temp.shape[0])
temp[temp['英雄名称']=='孙悟空']
```
