```python
#json --> javascript 对象json  
#是一种文本传输格式，前台和后台的传输格式
#处理cast以及crew这两列数据
#json.loads(str)-->obj
json_cols = ['cast','crew']
for i in json_cols:
    data2[i]=data2[i].apply(json.loads)
```

```python
#cast 提取演员名称
data2.cast[0]
def get_name(x):
    names = []
    for i in x:
        names.append(i['name'])
    return ",".join(names)
data2['cast']=data2.cast.apply(get_name)
```

```python
data2.crew[0]
```

```python
def get_director(x):
    for i in x:
        if i['job']=='Director':
            return i['name']
data2['crew'] = data2.crew.apply(get_director)
```

```python
#1.判断类型
# python 中的集合 
movie_types = set()
for i in data.genres.str.split(","):
    for item in i:
        movie_types.add(item)
movie_types.remove("")
```

```python
movie_types
```

```python
#                             genres                 Action Adventure Fantasy...
# 10000 Action,Adventure abc    action,Adventure        0       1        0   ...
```

```python
#for i in movie_types:
#    data[i] = data.genres.str.contains(i)
#data.genres.str.contains("Action")
#data = data.iloc[:,:-len(movie_types)]
```

```python
data['Action']=data.genres.str.contains("Action").apply(lambda x:1 if x else 0)
```

```python
data = data.drop(columns=['Action'])
```

```python
for i in list(movie_types):
    data[i]=data.genres.str.contains(i).apply(lambda x:1 if x else 0)
```

```python
data_types = data.loc[:,list(movie_types)]
```

```python
data_types
```

```python
#整理每一个电影是属于哪一年的
data.release_date
```

```python
# yyyy-mm-dd   yyyy/mm/dd    mm dd yy
data_types['year'] = pd.to_datetime(data.release_date,format='%Y-%m-%d').dt.year
```

```python
data_types
```

```python
#每一年电影的生产数量
result1 = data_types.groupby("year")['Western'].count()
```

```python
result1
```

```python
#电影数量随着年份的变化曲线
X = result1.index
```

```python
Y = result1.values
```

```python
plt.figure(figsize=(10,6))
plt.plot(X,Y,"-")
plt.xticks(np.arange(1915,2016,5))#,rotation=-45
plt.xlim([1915,2016])
```

```python
#每种电影随着年份的数量变换
result2 = data_types.groupby('year').sum().sort_index(ascending=False)
```

```python
result2.columns
```

```python
#result2['Action']
result3 = result2[['Animation','Action','Adventure','Crime','Comedy','Music']]
```

```python
plt.figure(figsize=(8,6))
plt.plot(result3,label=result3.columns)
plt.legend(result3)
plt.xticks(np.arange(1915,2016,5))#,rotation=-45
plt.xlim([1915,2016])
```

```python
#还可以使用柱状图去显示每个类型电影的数量
result4 = result2.sum()
```
