```python
import json# json模块是Python处理json数据的模块
```

```python
#json --> javascript 对象json  
#是一种文本传输格式，前台和后台的传输格式
#处理cast以及crew这两列数据
#json.loads(str)-->obj
json_cols = ['cast','crew']
for i in json_cols:
    data2[i]=data2[i].apply(json.loads)
```

```python
#cast 提取演员名称
data2.cast[0]
def get_name(x):
    names = []
    for i in x:
        names.append(i['name'])
    return ",".join(names)
data2['cast']=data2.cast.apply(get_name)
```

```python
data2.crew[0]
```

```python
def get_director(x):
    for i in x:
        if i['job']=='Director':
            return i['name']
data2['crew'] = data2.crew.apply(get_director)
```

```python
#修改列名
data2.rename(columns={'crew':'director'},inplace=True)
```
