```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def listplot():
    #根据数据绘制折线图
    plt.figure()#产生fig对象，axic对象，
    #X = np.arange(5) #[0,1,2,3,4]
    #Y = np.arange(5)+1 #[1,2,3,4,5]  =>(0,1),(1,2)....
    X = np.linspace(0,10,100)
    Y = np.sin(X)
    plt.plot(X,Y,'-')
```

```python
def listplot2():
    plt.figure(figsize=(6,4))
    X = np.arange(0,5,0.5)
    Y = X**2
    #x [...] y[...] -->[(x,y),(x,y)]
    #x [1,2,3],y[2,3,4]-zip->[(1,2),(2,3),(3,4)]
    plt.plot(X,Y,'-',color='red')   
    #plt.text(5,100,"important")
    #for i,j in zip(X,Y):
    #    plt.text(i+0.3,j,j,bbox=dict(facecolor='red', alpha=0.5))
    plt.title("这个是可视化")
    plt.xlabel("x轴标签")
    plt.ylabel("y轴标签")
    plt.xticks(np.arange(0,5,1),['a','b','c','d','e'])
listplot2()
```

```python
def listplot3():
    X = np.arange(50)
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'-',color='r',label='a')
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'--',color='y',label='b')
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'-*',color='c',label='c')
    plt.legend()#图示
listplot3()
```
