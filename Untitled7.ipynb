```python
import matplotlib.pyplot as plt
import numpy as np
#根据数据绘制折线图
plt.figure()#产生fig对象，axic对象，
#X = np.arange(5) #[0,1,2,3,4]
#Y = np.arange(5)+1 #[1,2,3,4,5]   =>(0,1),(1,2)...
X=np.linspace(0,10,10000)
Y=np.sin(X)
plt.plot(X,Y,'-')
```

```python
def listplot2():
    plt.figure(figsize=(16,12))
    X = np.arange(10)
    Y = X**2
    plt.plot(X,Y,'-o',color='#ff6600')
    for i,j in zip(X,Y):
        plt.text(i-0.2,j+2,j)
        plt.title("Title")
        plt.xlabel('x label')
        plt.ylabel('y label')
        plt.xticks(np.arange(0,10,0.5))
listplot2()
```

```python
def listplot3():
    plt.figure(figsize=(16,8))
    X = np.arange(50)
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'-',color='r',label='a')
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'-.',color='g',label='b')
    Y = np.random.randn(50).cumsum()
    plt.plot(X,Y,'--',color='b',label='c')
    plt.legend()
listplot3()
```

```python
def listplot4():
    X  = np.linspace(-np.pi,np.pi,256,endpoint = False)
    Y1,Y2 = np.sin(X),np.cos(X)
    plt.plot(X,Y1,'-')
    plt.plot(X,Y2,'-')
    plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'0',r'$\pi/2$',r'$\pi$'])
    #plt.legend()
    plt.ylim(-1,1)
    plt.xlim(-4,4)
    ax = plt.gca()#get current axic
    ax.spines["top"].set_color(None)
    ax.spines["right"].set_color(None)
    ax.spines['left'].set_position(('data',0))
    ax.spines['bottom'].set_position(('data',0))
    t = 2*np.pi/3
    y = np.sin(t)
    plt.scatter(t,y,50,color='r')
    plt.plot([t,t],[y,0],'--',color='g')
    #xycoords 表示当前的箭头的数据轴是基于整体数据的坐标轴
    plt.annotate('y=sin(x)',xy=(t,y),\
                 xycoords='data',textcoords='offset points',\
                xytext=(30,30))
listplot4()
```

```python
#柱状图
X=np.arange(4)
GDP=[2000,2345,2567,1890]
plt.bar(X,GDP)
plt.xticks(np.arange(4),['bj','sz','sh','dl'])
plt.ylim(1500,2900)
```

```python
#柱状图
Y=np.arange(4)
GDP=[2000,2345,2567,1890]
plt.barh(Y,GDP)
plt.yticks(np.arange(4),['bj','sz','sh','dl'])
plt.xlim(1500,2900)
```
