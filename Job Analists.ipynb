```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
data=pd.read_csv(r'C:\Users\29410\OneDrive\py\Files\job.csv',header=None)
del data[8]
data.columns = ['编号','职位','公司','工资下限','工资上限','地点','网址','就业方向']
```

```python
data.head(10)
```

```python
#每个城市工作岗位人数
data2=data.groupby('地点')['编号'].count().sort_values(ascending=False)
dataa2=[]
for i in data2.values:
    dataa2.append(math.log(i))
plt.figure(figsize=(60,4))
plt.bar(data2.index,dataa2)
plt.title("每个城市工作岗位人数")
plt.xlabel("城市")
plt.ylabel("工作数量（取ln对数）")
plt.figure(figsize=(12,4))
plt.bar(data2.index[0:20],dataa2[0:20])
plt.title("每个城市工作岗位人数（前20名）")
plt.xlabel("城市")
plt.ylabel("工作数量（取ln对数）")
print (data2.values[0:20])
for i,j,k in zip(np.arange(20),dataa2[0:20],data2.values[0:20]):
    plt.text(i-0.3,j+0.1,k)
data2
```

```python
#每个城市工资水平（平均值）
data['工资均值']=(data['工资下限']+data['工资上限'])/2
data3=data[data['工资均值']>0].groupby('地点')['工资均值'].mean().sort_values(ascending=False)
dataa3=[]
for i in data3.values:
    dataa3.append(math.log(i))
plt.figure(figsize=(60,4))
plt.bar(data3.index,data3)
plt.title("每个城市工资水平（平均值）")
plt.xlabel("城市")
plt.ylabel("工资水平")
plt.figure(figsize=(12,4))
plt.bar(data3.index[0:20],data3[0:20])
plt.title("每个城市工资水平（前20名）")
plt.xlabel("城市")
plt.ylabel("工资水平")
plt.figure(figsize=(12,4))
plt.title("每个城市工资水平（前20名）")
plt.xlabel("城市")
plt.ylabel("工资水平")
data3.head(20).plot(kind='bar')
for i,j in zip(np.arange(20),data3[0:20]):
    plt.text(i-0.4,j+500,int(j))
data3
```

```python
#每个城市每个公司招聘职位个数
data4=data.groupby('地点')['公司'].value_counts().sort_values(ascending=False)
data4
```

```python
#每个城市公司个数
data5=data.drop_duplicates(subset=['公司'],keep='first',inplace=False).groupby('地点')['公司'].count().sort_values(ascending=False)
dataa5=[]
for i in data5.values:
    dataa5.append(math.log(i))
plt.figure(figsize=(60,4))
plt.bar(data5.index,dataa5)
plt.title("每个城市公司个数")
plt.xlabel("城市")
plt.ylabel("工作数量（取ln对数）")
plt.figure(figsize=(12,4))
plt.bar(data5.index[0:20],dataa5[0:20])
plt.title("每个城市公司个数（前20名）")
plt.xlabel("城市")
plt.ylabel("工作数量（取ln对数）")
for i,j,k in zip(np.arange(20),dataa5[0:20],data5.values[0:20]):
    plt.text(i-0.3,j+0.1,k)
data5
```

```python
#自我对图像方面的研究方向比较问兴趣，筛选符合条件的招聘条目，重复上面步骤。
mydata = pd.DataFrame(columns =data.columns)
for i in range(data.shape[0]):
    if (data.iloc[i])['就业方向']==(data.iloc[i])['就业方向'] \
    and (('图像' in (data.iloc[i])['就业方向']) \
    or ('人脸' in (data.iloc[i])['就业方向']) \
    or ('视觉' in (data.iloc[i])['就业方向'])):
        mydata.loc[i]=data.iloc[i]
mydata.head(20)
```

```python
#每个城市工作岗位人数
data2=mydata.groupby('地点')['编号'].count().sort_values(ascending=False)
dataa2=[]
for i in data2.values:
    dataa2.append(math.log(i))
plt.figure(figsize=(60,4))
plt.bar(data2.index,dataa2)
plt.title("每个城市工作岗位人数")
plt.xlabel("城市")
plt.ylabel("工作数量（取ln对数）")
plt.figure(figsize=(12,4))
plt.bar(data2.index[0:20],dataa2[0:20])
plt.title("每个城市工作岗位人数（前20名）")
plt.xlabel("城市")
plt.ylabel("工作数量（取ln对数）")
print (data2.values[0:20])
for i,j,k in zip(np.arange(20),dataa2[0:20],data2.values[0:20]):
    plt.text(i-0.3,j+0.1,k)
data2
```

```python
#每个城市工资水平（平均值）
mydata['工资均值']=(mydata['工资下限']+mydata['工资上限'])/2
data3=mydata[mydata['工资均值']>0].groupby('地点')['工资均值'].mean().sort_values(ascending=False)
dataa3=[]
for i in data3.values:
    dataa3.append(math.log(i))
plt.figure(figsize=(60,4))
plt.bar(data3.index,data3)
plt.title("每个城市工资水平（平均值）")
plt.xlabel("城市")
plt.ylabel("工资水平")
plt.figure(figsize=(12,4))
plt.title("每个城市工资水平（前20名）")
plt.xlabel("城市")
plt.ylabel("工资水平")
data3.head(20).plot(kind='bar')
for i,j in zip(np.arange(20),data3[0:20]):
    plt.text(i-0.4,j+500,int(j))
```

```python
#每个城市每个公司招聘职位个数,可以看出由于数据量大，有一些公司省略了
data4=mydata.groupby('地点')['公司'].value_counts().sort_values(ascending=False)
data4
```
